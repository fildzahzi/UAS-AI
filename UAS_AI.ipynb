```python
#breast cancer dataset : classification models
#import modules
from sklearn.model_selection import train_test_split #pembagi dataset menjadi training dan testing test
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report #evaluasi performa model
```

```python
#dataset : features & class label
X = cancer_df.drop(columns='diagnosis') #menempatkan features ke dalam variable X
X.head() #tampilkan 5 baris pertama
```

```python
y = cancer_df['diagnosis'] #menempatkan class label (target) ke dalam variable y
y.head() #menampilkan 5 baris pertama
```

```python
#Split the dataset into a training set and testing test
#membagi dataset kedalam training dan testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)

print('training dataset')
print(X_train.shape)
print(y_train.shape)
print()
print('testing dataset :')
print(X_test.shape)
print(y_test.shape)
```

```python
#K-Nearest Neighbors Algorithm

#K-Nearest Neighbors(KNN)
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_range = list(range(1,26))
scores = []
for k in k_range:
  model_knn = KNeighborsClassifier(n_neighbors=k) #konfigurasi algoritma
  model_knn.fit(X_train, y_train) #melakukan prediksi
  y_pred = model_knn.predict(X_test) #melakukan prediksi
  scores.append(accuracy_score(y_test, y_pred)) #evaluasi peforma
```

```python
plt.plot(k_range, scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accurancy Score')
plt.title('Accurancy Scores for Values of k of k-Nearest-Neighbors')
plt.tight_layout()
plt.show()
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3) #konfigurasi algoritma
model_knn.fit(X_train,y_train) #training model / classifier
y_pred = model_knn.predict(X_test) #melakukan prediksi
```

```python
#accurancy score
print(accuracy_score(y_test, y_pred)) #evaluasi akurasi
```

```python
#confussi matrix
print(confusion_matrix(y_test, y_pred)) #evaluasi confusion matrix
```

```python
#Classification Report
print(classification_report(y_test, y_pred)) #evaluasi klasifikasi
```
